```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.tri as mtri

# Veri setini oluşturalım (örnek veri)
age = np.linspace(20, 80, num=50)
weight = np.random.uniform(50, 100, size=50)

# Yeni bir figür oluşturalım
fig = plt.figure(figsize=plt.figaspect(0.5))

# ==========
# First plot (Mobius mapping)
# ==========

# Make a mesh in the space of parameterisation variables u and v
u = np.linspace(0, 2.0 * np.pi, endpoint=True, num=50)
v = np.linspace(-0.5, 0.5, endpoint=True, num=10)
u, v = np.meshgrid(u, v)
u, v = u.flatten(), v.flatten()

# This is the Mobius mapping, taking a u, v pair and returning an x, y, z
# triple
x1 = (1 + 0.5 * v * np.cos(u / 2.0)) * np.cos(u)
y1 = (1 + 0.5 * v * np.cos(u / 2.0)) * np.sin(u)
z1 = 0.5 * v * np.sin(u / 2.0)

# Triangulate parameter space to determine the triangles
tri1 = mtri.Triangulation(u, v)

# Plot the surface.  The triangles in parameter space determine which x, y, z
# points are connected by an edge.
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.plot_trisurf(x1, y1, z1, triangles=tri1.triangles, cmap=plt.cm.Spectral)
ax1.set_zlim(-1, 1)
ax1.set_title('Mobius Mapping')

# ===========
# Second plot (Age and Weight)
# ===========

# Create x, y, z based on age and weight data
x2 = age
y2 = weight
z2 = np.zeros_like(x2)  # Dummy z values (all zeros)

# Create the Triangulation
tri2 = mtri.Triangulation(x2, y2)

# Plot the surface
ax2 = fig.add_subplot(1, 2, 2, projection='3d')
ax2.plot_trisurf(tri2, z2, cmap=plt.cm.CMRmap)
ax2.set_title('Age and Weight')

# Görseli ekranda göster
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Metin verisi
text_data = "veri bilimi yapay zeka çok iyi olacak teknoloji gibi kelimeleri elon musk jeff bezoss turkiye hamza gündüz sude toprak health medicine doctor hospital treatment surgery diagnosis therapy wellness patient nutrition exercise healthcare recovery prevention vaccination mental health fitness"

# Kelime bulutu oluşturma fonksiyonu
def generate_wordcloud(text):
    """Kelime bulutu oluşturma fonksiyonu."""
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

if __name__ == "__main__":
    generate_wordcloud(text_data)
```
